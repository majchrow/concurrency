```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dfs = [pd.read_csv(mode + str(number), sep=" ", names=[mode, number]) 
       for mode in ("waiter", "beb", "asym", "naive") for number in range(3,16)]    
```

```python
dfs_with_mean = [df.groupby(df.columns[0]).mean() for df in dfs]
```

```python
for df in dfs_with_mean:
    df.plot.bar()
```
